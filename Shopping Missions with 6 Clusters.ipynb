```python
import pandas as pd
import numpy as np

pos_cluster = pd.read_excel('pos_cluster_v2.xlsx', sheet_name='Sheet1')
pos_cluster

xls = pd.ExcelFile('C:\\Users\\MOCCA\\Desktop\\RFManalysis\\my_project\\6clusters.xlsx')

def shopMission(xls, nameClust):
    cluster = pd.read_excel(xls, nameClust)
    cluster_merged = pd.merge(pos_cluster, cluster, how="right", on="Basket ID")
    cluster_merged_count = cluster_merged.groupby(['CategoryC', 'CategoryCDescription']).size().reset_index(name= 'Count')
    cluster_merged_count.sort_values(by="Count", ascending=False, inplace=True)
    return cluster_merged_count;
```
