```python
import pandas as pd
import numpy as np

pos_cluster = pd.read_excel('pos_cluster_v2.xlsx', sheet_name='Sheet1')
pos_cluster

xls = pd.ExcelFile('cluster_resluts.xlsx')
cluster1 = pd.read_excel(xls, 'Cluster 1')
```

```python
cluster1_merged = pd.merge(pos_cluster, cluster1, how="right", on="Basket ID")
cluster1_merged
cluster1_merged_count = cluster1_merged.groupby(['CategoryC', 'CategoryCDescription']).size().reset_index(name = 'Count')
cluster1_merged_count.sort_values(by="Count", ascending=False, inplace=True)
cluster1_merged_count
#EIDI UGEIINIS
```
